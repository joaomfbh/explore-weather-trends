### Data Wrangle

```python
df = pd.read_csv('weather_trends.csv')
```

```python
df.info()
```

```python
# 7 years moving average
df['mva_local'] = df['avg_temp'].rolling(7).mean()
```

```python
# 7 years moving average
df['mva_global'] = df['avg_temp_global'].rolling(7).mean()
```
